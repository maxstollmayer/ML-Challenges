### Data Cleanup
We will now remove rows that have data points that do not lie in the accepted range of values.

```python
binaryCols = ["Running", "Blue Switch On", "User had coffee", "Controller lightgoldenrodyellow", "Controller lightgray",
              "Controller lightgreen", "Cruise control", "Voltage main", "Voltage secondary", "Counterweight",
              "Second Counterweight", "Smoke formation", "Bending of test plate"]
nonnegativeCols = ["Time since restart"]
intervalCols = givenDF.columns.difference(binaryCols).difference(nonnegativeCols).difference(["CPU Temperature"])

for col in binaryCols:
    dropIndices = givenDF[(givenDF[col] != 0) & (givenDF[col] != 1)].index

for col in nonnegativeCols:
    dropIndices.append(givenDF[givenDF[col] < 0].index)

for col in intervalCols:
    dropIndices.append(givenDF[(givenDF[col] > 1) | (givenDF[col] < 0)].index)
    
cleanDF = givenDF.drop(index = dropIndices).drop_duplicates().reset_index(drop=True)
cleanDF.describe()
```

Again we plot the mean, standard deviation, minimum and maximum by column and as a histogram.

```python
cleanGroup = cleanDF.groupby("Running")

fig1, axs1 = plt.subplots(2, 2)

cleanGroup.mean().T.plot(ax=axs1[0,0], legend=False, xticks=[])
cleanGroup.mean().T.plot(kind="hist", ax=axs1[0,1], legend=False)
axs1[0,0].set_ylabel("Mean")
axs1[0,0].set_title("Per columns")
axs1[0,1].set_ylabel("")
axs1[0,1].set_title("Histograms")

cleanGroup.std().T.plot(ax=axs1[1,0], legend=False, xticks=[])
cleanGroup.std().T.plot(kind="hist", ax=axs1[1,1], legend=False)
axs1[1,0].set_ylabel("Standard deviation")
axs1[1,1].set_ylabel("")

fig1.legend(labels = ["Running", "Not running"], bbox_to_anchor=(0.25, 0.875))
plt.show()


fig2, axs2 = plt.subplots(2, 2)

cleanGroup.min().T.plot(ax=axs2[0,0], legend=False, xticks=[])
cleanGroup.min().T.plot(kind="hist", ax=axs2[0,1], legend=False)
axs2[0,0].set_ylabel("Minimum")
axs2[0,1].set_ylabel("")

cleanGroup.max().T.plot(ax=axs2[1,0], legend=False, xticks=[])
cleanGroup.max().T.plot(kind="hist", ax=axs2[1,1], legend=False)
axs2[1,0].set_ylabel("Maximum")
axs2[1,1].set_ylabel("")

plt.show()
```

### Balanced Training and Test Set
The number of 'Running' and 'Not running' labels are not equal. To not include a bias in the classifier for either label we balance the data set by oversampling the 'Running' entries.

```python
cleanDF["Running"].value_counts().plot(kind="pie", labels=["Not running", "Running"], label="", autopct="%0.1f%%")
plt.show()
```

```python
copyAmount = cleanDF.shape[0] - 2 * cleanDF[cleanDF["Running"] == 1].shape[0]
balancedDF = cleanDF.append(cleanDF[cleanDF["Running"] == 1].sample(copyAmount, replace=True)).reset_index(drop=True)

balancedDF["Running"].value_counts().plot(kind="pie", labels=["Not running", "Running"], label="", autopct="%0.1f%%")
plt.show()
```

To train our classifiers we split the balanced set into the target labels and the training data.

```python
trainData = balancedDF.iloc[:, 1:].values
trainLabels = balancedDF["Running"].values

np.savetxt("train_data.csv", trainData, delimiter=",")
np.savetxt("train_labels.csv", trainLabels, delimiter=",")
```

### Simple Classifier
To get a feel for the dependencies between the columns we plot the correlation matrix and the correlation of every other column with 'Running'.

```python
sn.heatmap(cleanDF.corr(), cmap="viridis", square=True, vmin=-1, vmax=1)
plt.title("Correlation Matrix")
plt.show()

cleanDF.corr()["Running"][1:].plot()
plt.title("Correlation with 'Running'")
plt.show()
```

Three of the columns are very correlated, they are:

```python
corrCols = cleanDF.loc[:, cleanDF.corr()["Running"].abs() > 0.2].columns[1:]
cleanDF.corr()["Running"][corrCols]
```

Plotting the confusion matrix with the binary column 'Blue Switch On' we see that if the switch is off then the machine cannot run.

```python
confusionMatrix = pd.crosstab(cleanDF["Running"].astype(bool), cleanDF["Blue Switch On"].astype(bool))
sn.heatmap(confusionMatrix[::-1], cmap="viridis", annot=True, square=True, fmt="d")
plt.title("Confusion Matrix")
plt.show()
```

In our simple classifier we will use the observation above as our first condition for labeling as 'Running' and a threshold for the difference between 'Controller mistyrose' and 'Controller moccasin' since the second one is negatively correlated. The best threshold is found by iterating through the training data and noting for which threshold the error is smallest.

```python
class SimpleClassifier:
    def __init__(self, bestThreshold=None):
        self.bestThreshold = bestThreshold
    
    def predict(self, data, threshold=None):
        threshold = threshold if threshold is not None else self.bestThreshold
        if threshold is None:
            raise Exception("Not fitted yet!")
            
        switchOn = data[:, 0] == 1
        controllerDiff = data[:, -2] - data[:, -1] >= threshold
        predictedLabels = switchOn & controllerDiff
        return predictedLabels
    
    def fit(self, data, labels, meshSize=1001):
        bestError = data.shape[0]
        for threshold in np.linspace(-1, 1, meshSize):
            error = (self.predict(data, threshold=threshold) != labels).sum()
            if error < bestError:
                bestError = error
                self.bestThreshold = threshold
    
    def score(self, data, labels):
        return 1 - (self.predict(data) != labels).sum() / data.shape[0]
    
    def get_params(self, deep):
        # for compatability with sklearn
        return {"bestThreshold": self.bestThreshold}
```

## Model Selection
The simple classifier from above is compared to standard classifiers from the scikit-learn library by using a cross-validation routine on the training set.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB

rng = np.random.RandomState(42) # locks the randomness

classifiers = {
    "Simple":            SimpleClassifier(),
    "Nearest Neighbors": KNeighborsClassifier(),
    "linear SVM":        SVC(kernel="linear", random_state=rng),
    "polynomial SVM":    SVC(kernel="poly", random_state=rng),
    "radial SVM":        SVC(kernel="rbf", random_state=rng),
    "sigmoid SVM":       SVC(kernel="sigmoid", random_state=rng),
    "Gaussian Process":  GaussianProcessClassifier(random_state=rng),
    "Decision Tree":     DecisionTreeClassifier(random_state=rng),
    "Random Forest":     RandomForestClassifier(random_state=rng),
    "AdaBoost":          AdaBoostClassifier(random_state=rng),
    "Neural Net":        MLPClassifier(alpha=1, max_iter=1000, random_state=rng),
    "Naive Bayes":       GaussianNB()
}
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold

cvSplit = StratifiedKFold(10, shuffle=True, random_state=rng)
comparisonTable = pd.DataFrame()

for name, clf in classifiers.items():
    cvScores = cross_val_score(clf, trainData, trainLabels, cv=cvSplit, scoring="accuracy")
    comparisonTable.loc[name, "Mean CV score"] = cvScores.mean()
    comparisonTable.loc[name, "Standard deviation"] = cvScores.std()
    
comparisonTable.style.highlight_max("Mean CV score")
```

The decision tree, random forest and AdaBoost usually perform the best for the cross-validation on our training data but our simple classifier also achieves a very good score.

### Hyperparameter Tuning
To find the optimal hyperparameter of the best classifiers from above we use a cross-validation routine on the training data.

```python
from sklearn.model_selection import GridSearchCV

tuning = {
    "Decision Tree": (DecisionTreeClassifier(random_state=rng),
                      {"min_samples_leaf": [1, 2, 5, 10]}),
    "Random Forest": (RandomForestClassifier(random_state=rng),
                      {"n_estimators": [10, 50, 100], "min_samples_leaf": [1, 2, 5, 10]}),
    "AdaBoost":      (AdaBoostClassifier(random_state=rng),
                      {"n_estimators": [10, 50, 100],
                       "base_estimator": [DecisionTreeClassifier(random_state=rng),
                                          RandomForestClassifier(random_state=rng)]})
}

tuningTable = pd.DataFrame()
bestEstimators = []

for name, (clf, parameter) in tuning.items():
    gridClf = GridSearchCV(clf, parameter, cv=cvSplit)
    gridClf.fit(trainData, trainLabels)
    bestClf = gridClf.best_estimator_
    bestEstimators.append((name, bestClf))
    tuningTable.loc[name, "Mean CV score"] = gridClf.best_score_
    tuningTable.loc[name, "Standard deviation"] = gridClf.cv_results_["std_test_score"][gridClf.best_index_]
    tuningTable.loc[name, "Best parameters"] = str(gridClf.best_params_)
    
tuningTable.style.highlight_max("Mean CV score")
```

### Ensemble Model
To gain the benefit of the three best models we can use an ensemble model. Here we chose a stacking classifier using a logistic regression as the final estimator.

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression

stackClf = StackingClassifier(estimators=bestEstimators, final_estimator=LogisticRegression(random_state=rng))
print("Mean CV score:", cross_val_score(stackClf, trainData, trainLabels, cv=cvSplit, scoring="accuracy").mean())
```

## Our Classifier
Finally the classifier we submit is defined below as the stacking classifier from above and with outlier rows being labeled as 'Not running'.

```python
class CombinedClassifier:
    def __init__(self, classifier):
        self.classifier = classifier
        
    def predict(self, data):        
        # prediction of classifier
        prediction = self.classifier.predict(data)

        # detect outliers and set label to 0
        finalPrediction = [prediction[index] if row.std() < 15 else 0 for index, row in enumerate(data)]
        return np.array(finalPrediction).astype(int)
        
    def fit(self, data, labels):
        self.classifier.fit(data, labels)
        
    def score(self, data, labels):
        return 1 - (self.predict(data) != labels).sum() / data.shape[0]
    
    # for combatability with sklearn
    def get_params(self, deep):
        return {"classifier": self.classifier}
```

```python
from joblib import dump, load

ourClassifier = CombinedClassifier(stackClf)
print("Mean CV score:", cross_val_score(ourClassifier, trainData, trainLabels, cv=cvSplit, scoring="accuracy").mean())

ourClassifier.fit(trainData, trainLabels)
dump(ourClassifier, "ourClassifier.joblib")

ourClassifier = load("ourClassifier.joblib")
```
